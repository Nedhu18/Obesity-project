```python
num_obse = ['Weight', 'FAF', 'CH2O', 'TUE']
obfig, obaxes = pltobsty.subplots(ncols=1, nrows=len(num_obse), figsize=(14, 6 * len(num_obse)))

for iobst, obf in enumerate(num_obse):
    snsobsty.boxplot(data=dfobesty, x='Obesity', y=obf, ax=obaxes[iobst], palette='coolwarm')
    obaxes[iobst].set_title(f'{obf} by Obesity Level', fontsize=14, fontweight='bold')
    obaxes[iobst].set_xlabel('Obesity Level', fontsize=12)
    obaxes[iobst].set_ylabel(obf, fontsize=12)
pltobsty.tight_layout()
pltobsty.show()
```

```python
obstfig, axob = pltobsty.subplots(3, 2, figsize=(14, 11))

# family_history Across Obesity
snsobsty.countplot(x="family_history", data=dfobesty, hue="Obesity",ax=axob[0, 0])
axob[0, 0].set_title("Impact of family history of overweight on Obesity")

# FAVC Across Obesity
snsobsty.countplot(x="FAVC", hue="Obesity", data=dfobesty, ax=axob[0, 1])
axob[0, 1].set_title("Impact of Frequent High-Calorie Food Intake on Obesity")
axob[0, 1].tick_params(rotation=42, axis='x')

# CAEC Across Obesity
snsobsty.countplot(x="CAEC", hue="Obesity", data=dfobesty, ax=axob[1, 0])
axob[1, 0].set_title("Impact of Between-Meal Snacking on Obesity")

# Smoke Across Obesity
snsobsty.countplot(x="SMOKE", hue="Obesity", data=dfobesty, ax=axob[1, 1])
axob[1, 1].set_title("SMOKE vs Obesity")

# CALC Across Obesity
snsobsty.countplot(x="CALC", hue="Obesity", data=dfobesty, ax=axob[2, 0])
axob[2, 0].set_title("CALC vs Obesity")

# MTRANS Across Obesity
snsobsty.countplot(x="MTRANS", hue="Obesity", data=dfobesty, ax=axob[2, 1])
axob[2, 1].set_title("MTRANS vs Obesity")

pltobsty.tight_layout()
pltobsty.show()
```

converting all the columns with object datatype to numerical

```python
from sklearn.preprocessing import LabelEncoder as lbodsty
```

```python
oblbb = lbodsty()
cob = dfobesty.select_dtypes(include=['object']).columns
for column in cob:
    dfobesty[column] = oblbb.fit_transform(dfobesty[column])
```
